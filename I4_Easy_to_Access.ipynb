# Example #1: file download and read locally

```python
from google.colab import files
import pandas as pd

# Upload CSV file
uploaded = files.upload()

# Read the CSV file, assuming only one file is uploaded
file_name = list(uploaded.keys())[0]
df = pd.read_csv(file_name)

# show sample data
df.head()
```

The code above allows us to upload and read a local csv file. For this particular example, it reads a CSV dataset called "Programming Languages Trend Over Time" originally downloaded from Kaggle:
https://www.kaggle.com/datasets/nextmillionaire/programming-languages-trend-over-time?resource=download

*   Pros of the access methodology: Accessing local data usually gives us more control over how and what data we want to access. Also, it has less dependency on the external factors such as internet connection.
*   Cons of the access methodology: Some data takes a lot of space, so it is somtimes impossible to store them locally. Even if the data can be stored locally, it can be a challenge to maintain it overtime.

# Example #2: API connection

```python
import requests

# request API connection over HTTP request to the a site called Randomfox
response = requests.get("https://randomfox.ca/floof")
# parses the response in json format
fox = response.json()
# show a sample data: a fox picture
print(fox['image'])
```


*   Pros of the access methodology: API connection has a high advantage eal-time or near-real-time data updates over other methods. Alao, APIs typically provide data in a structured format like JSON or XML, which makes it easier to parse and integrate data.
*   Cons of the access methodology: Becasue the availability of data is highly dependent on the API connection, reliability of the API matters. API connection is not suitable for sentitive data.

# Example #3: HTTP to access webpage

```python
# URL of the online bookï¼šTHE STAR-STEALERS By EDMOND HAMILTON
url = "https://www.gutenberg.org/cache/epub/73442/pg73442-images.html"
# send an HTTP request to get HTML content
response = requests.get(url)
# show a sample of the text
print(response.text[10000:10200])
```


*   Pros of the access methodology: Using HTTP is relatively simple and lightweight, so that users can eaiser publish and spread data over the website.
*   Pros of the access methodology: HTTP also has risks of data corruption and security issues. It does not guarantee users to get the data at all occasions, and users usually do not know when a web changes.
